```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings('ignore')
sns.set_style("darkgrid")
```

```python
def get_gene_names():
    return [file_name.split('.')[0].split('_')[-1] for file_name in os.listdir("../data/raw/") if file_name.endswith('.fasta')]    
```

```python
def report_on_gene(gene:str):
    gene = gene.upper()
    print(f"Report on {gene}")
    df = pd.read_csv(f'../data/raw/felidae_{gene}_info.csv')
    df = df.dropna()
    display(df.head())
    display(df.slen.describe())
    
    # generating plots
    fig, ax = plt.subplots(1, 2, figsize=(12,6))
    sns.histplot(x=df.slen, ax=ax[0])
    ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45, ha="right")
    sns.boxplot(y=df.slen, ax=ax[1])
    plt.show()
    
```

```python
genes = get_gene_names()
for gene in genes:
    report_on_gene(gene)
```
